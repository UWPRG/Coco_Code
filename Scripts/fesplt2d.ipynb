```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ploting a 2D Free energy surface
# X axis is distance form surface y axis is coordination number
N = 80
M = 164
subplot = 0
plt.figure(figsize=(16,6))
for i in range(10):
    data = np.genfromtxt('fesfiles/fes_{}.dat'.format(i),comments='#')
    X = data[:,0].reshape((N,M))
    Y = data[:,1].reshape((N,M))
    Z = data[:,2].reshape((N,M))
    spacer = -0.4
    lines = 10
    levels=np.linspace(0,lines*spacer,num=(lines+1),endpoint=True)
# chose FES to plot    
    if (i+1) in (8,9,10):
        subplot += 1
        plt.subplot(1,3,subplot)
        
        plt.contourf(X, Y, Z, cmap=plt.cm.inferno)
        plt.colorbar()
        plt.xlabel('Distance from surface')
        plt.ylabel('Comformation')
        #plt.ylim([1.5,6])
        #plt.xlim([1.5,6])
        plt.title('{}0%'.format(i+1))
        plt.rcParams.update({'font.size': 8})
plt.tight_layout()
plt.show()
```
